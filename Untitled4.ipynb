```python
import pandas as pd
import geopandas as gpd
```

```python
ambulance=pd.read_csv('allshortam1.csv')
police=pd.read_csv('allshortpo1.csv')
both=[ambulance,police]
both=pd.concat(both)
ambulance=ambulance[ambulance.incident_type.str.contains("OVERDOSE")]
ambulance=ambulance[ambulance.Location.duplicated(keep=False)]
ambulance=ambulance[ambulance.Response_Time.notnull()]
police=police[police.incident_type.str.contains("OVERDOSE")]
police=police[police.Location.duplicated(keep=False)]
police=police[police.Response_Time.notnull()]
both=both[both.incident_type.str.contains("OVERDOSE")]
both=both[both.Location.duplicated(keep=False)]
both=both[both.Response_Time.notnull()]
both.columns=both.columns.str.lower()
both.towns=both.towns.str.lower()
both=both[both.towns.notnull()]
odtowns=['bristol township','bensalem township','falls township','middletown township','warminster township']
closer=both[both['towns'].isin(odtowns)]
closer['dispatched']=pd.to_datetime(closer.dispatched)
top_towns=closer[closer.dispatched.dt.year>=2016]
ems_map=gpd.read_file('C:\\Users\\ulleryc\\data-newsroom\\mapping_data\\fwemailing2490_ullery_911_request_pdf\\EMS_Boxes.shp')
ems_map.rename(columns={'EMS_BLOCK_':'ems_blk'},inplace=True)
ems_map['ems_blk']=ems_map['ems_blk'].astype(str)
top_towns['ems_blk']=top_towns['ems_blk'].astype(str)
top_towns['ems_blk']=top_towns['ems_blk'].str[:6]
```
